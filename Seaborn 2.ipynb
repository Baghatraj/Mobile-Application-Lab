```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.info()
```

## Correlation with heatmap

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

## Jointplot

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='reg')
```

## PairPlot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

## Distplot

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.distplot(iris['petal_length'],kde=False)
```

# Categorcial Plots

## Count Plot

```python
sns.countplot('species',data=iris)
```

## Bar Plot

```python
sns.barplot(x='species',y='petal_length',data=iris)
```

```python
sns.barplot(x='species',y='sepal_length',data=iris)
```

```python
sns.barplot(x='species',y='petal_width',data=iris)
```

```python
sns.barplot(x='species',y='sepal_width',data=iris)
```

```python
sns.barplot(data=iris)
```

## Box Plot

```python
sns.boxplot(x='species',y='petal_length',data=iris)
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
```

```python
sns.boxplot(data=iris)
```

## Violin Plot

```python
sns.violinplot('species','sepal_length',data=iris)
```

```python
sns.violinplot(data=iris)
```
