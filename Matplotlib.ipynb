```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple graph')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
```

```python
x = np.arange(0 ,4*np.pi,0.1)
y=np.sin(x)

plt.plot(x,y,'r')
```

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,4,7]
y2=[6,17,20]
plt.bar(x,y)
plt.bar(x2,y2)
plt.show()
```

```python
a = np.random.randint(0,100,10)
plt.hist(a)
plt.show()
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
labels = 'py','c','c++','java'
sizes = [27,33,18,22]
colors = ['red','yellow','green','black']
explode = [0,0.1,0,0]

plt.pie(sizes,explode=explode,labels=labels,colors=colors)
plt.show()
```
